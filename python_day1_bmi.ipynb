```python
is_weekend = True
has_homework = False

if is_weekend and not has_homework:
  print("푹 쉴 수 있어요")
elif is_weekend and has_homework:
  print("숙제를 먼저 해요")
elif not is_weekend and has_homework:
  print("오늘도 바빠요")
else:
  print("조금 여유롭네요")
```
